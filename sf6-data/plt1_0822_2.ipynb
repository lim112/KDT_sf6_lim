#### 시각화 라이브러리 사용 matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
# 리스트를 한개만 사용시 y축 값으로 인식야여 작성된다
# plt.plot([1,2,3,4],[10,30,20,40])
        # 가로축 값, 세로축 값
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Basic Graph')
plt.plot([1,2,3,4],[90,80,85,70])
plt.xlabel('학생번호')
plt.ylabel('시험점수')
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('범례 색상 표시')
plt.plot([10,20,30,40], label = '체중',color='g')
plt.plot([40,30,20,10], label = '출산률',color ='r')
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('범례 색상 표시')
plt.plot([10,20,30,40],'b.', label = 'circle')
                        # 색과 모양
plt.plot([40,30,20,10],'r^', label = 'triangle')
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.legend()
plt.show()
```
