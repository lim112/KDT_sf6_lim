```python
# !pip install folium

import folium

# 지도 생성
seoul_map = folium.Map(location=[37.539705, 126.94604], zoom_start=15)

# 위치 리스트
locations = [
    [37.539705, 126.94604],  # 마포
    [37.497952, 127.027619],  # 강남역
    [37.571731, 127.011069]   # 동대문
]
# 리스트 딕셔너리
map_info = [
    {'location' : [37.539705, 126.94604], 'popup' : '마포역'},
    { 'location' : [37.497952, 127.027619], 'popup' : '강남역'},
     {'location' : [37.571731, 127.011069], 'popup' : '동대문역'}
]
# 각 위치에 마커 추가
for map in map_info:
    folium.Marker(
        location=map['location'],
        popup = map['popup'],
        icon=folium.Icon(icon='star', color='blue')
    ).add_to(seoul_map)

# 지도 출력
seoul_map
```
