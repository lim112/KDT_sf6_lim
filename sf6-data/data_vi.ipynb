```python
import csv
import matplotlib.pyplot as plt

with open("culture_a.csv", "r") as f:
    data = csv.reader(f) 
    header = next(data)

    x_labels = []  # x축 레이블(지역 이름)
    y_values = []  # y축 값(빈도)

    for row in data:
        if row[-1] !='':
            if row[0] == '연령별':
                x_labels.append(row[1])
                for index_header, value in enumerate(header):
                    if '2014년' in value.split():
                        if '전시회' in value.split():
                            y_values.append(float(row[index_header].replace(',', '')))
                
            
plt.rc('font', family='Malgun Gothic')
plt.title("서울시의 2014년의 연령별 문화생활 평균소비")
plt.bar(x=x_labels, height=y_values)
plt.xlabel('연병별')
plt.ylabel('평균소비 금액')
plt.xticks(rotation=45)  # x축 레이블이 겹치지 않도록 회전
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

with open("culture_b.csv", "r") as f:
    data = csv.reader(f) 
    header = next(data)

    x_labels = []  # x축 레이블(지역 이름)
    y_values = []  # y축 값(빈도)

    for row in data:
        if row[-1] !='':
            if row[0] == '연령별':
                x_labels.append(row[1])
                for index_header, value in enumerate(header):
                    if '2014년' in value.split():
                        if '전시회' in value.split():
                            y_values.append(float(row[index_header].replace('-', '0')))
                
            
plt.rc('font', family='Malgun Gothic')
plt.title("서울시의 2014년의 연령별 문화생활 빈도")
plt.bar(x=x_labels, height=y_values)
plt.xlabel('연령별')
plt.ylabel('빈도')
plt.xticks(rotation=45)  # x축 레이블이 겹치지 않도록 회전
plt.show()
```
