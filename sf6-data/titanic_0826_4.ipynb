```python
# 누락 데이터 확인 - isnull() - True(누락), False(누락 ㄴ) , notnull()
df.head().isnull()

# 누락 데이터 총 개수 계산
df.head().isnull().sum()
df.isnull().sum()

# value_counts()
# print(df['deck'].value_counts(dropna = False))
# 누락된 데이터 확인 NaN

print(df['deck'].value_counts())
print(df['age'].value_counts())

# age행의 누락데이터 확인
df['age'].head()
```

```python
# 누락 데이터 삭제 - nan값이 500개 이상인 열 삭제, dropna() 메서드 사용
# thresh = 500지정 , axis = 1
df_thresh = df.dropna(axis =1, thresh = 500)
# 삭제 후 남아있는 칼럼
# print(df_thresh.columns)

# age열에 데이터가 없는 모든 행 삭제
# subset을 'age'열로 한정하면 'age'열의 행 중에서 NaN값이 있는 모든 행 삭제
df_age = df.dropna(subset = 'age', how = 'any', axis = 0 )
# 삭제 후 행의 개수
print(len(df_age))
# print(df['age'].heda[10])
print(df_age.head(10))
```

```python

import seaborn as sns
df = sns.load_dataset('titanic')
df

print(df['age'].head(10))
# df.info()

# 누락된 데이터 채우기 fillna()
# 1 평균(mnan()) 
mean_age = df['age'].mean(axis = 0)
mean_age = df['age'].fillna(mean_age)
print(mean_age.head(10))

# 2 최빈값(mode())
# 승선한 도시의 NaN을 최빈값으로 대체
df['embark_town'][820:831]
most_freq = df['embark_town'].mode()[0]
df['embark_town'] = df['embark_town'].fillna(most_freq)
print(df['embark_town'][820:830])

```
