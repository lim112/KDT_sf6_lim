```python
import folium

# 지도 객체 생성
seoul_map = folium.Map(location=[37.54322, 126.951576], zoom_start=15)

# Marker 클래스를 사용하여 마커 추가
folium.Marker(
    location=[37.54322, 126.951576],
    icon=folium.Icon(icon='star', color='red')
).add_to(seoul_map)

# CircleMarker 클래스를 사용하여 원형 마커 추가
folium.CircleMarker(
    location=[37.54322, 126.951576],
    radius=100,  # 반지름 설정
    color='#ff0',  # 색상 설정
    fill=True,  # 원 내부 채우기
    fill_color='#ff0'
).add_to(seoul_map)

# 지도를 HTML로 저장
seoul_map.save('./datas/seoul.html')

# 결과 확인 
seoul_map
```
