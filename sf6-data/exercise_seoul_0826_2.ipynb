```python
import pandas as pd
import warnings 

warnings.simplefilter('ignore')
# 없이하면 경고창이 생성됨
# excel 파일 불러오기 - read_excel(경로)
not_exercise = pd.read_excel('./datas/exercise_soeul.xlsx', header = 1)
                                                            # 제목행부터 출력
# not_exercise

# 시점 열 삭제 drop() , inplace = True 저장
not_exercise.drop(columns = '시점' , inplace = True)
# not_exercise.drop(rows = '' , inplace = True)


# 칼럼 이름 변경 .rename() - 구분별(1) > 대분류, 구분별(2) > 분류
not_exercise.rename(columns = {'구분별(1)' : '대분류','구분별(2)' : '분류'}, inplace = True)
not_exercise

# 22행 ~ 51행 삭제
not_exercise.drop(index = range(22,52), inplace = True)
print(not_exercise)
```

```python
# NaN변경
not_exercise.loc[2, "대분류"] = '성별'
# print(not_exercise)

# 성별 데이터만 선택
not_exercise.loc[2, "대분류"] = '성별'
not_ex_age = not_exercise
not_exercise.loc[4:8, "대분류"] = '연령별'
                    # 여러 데이
not_ex_age = not_exercise[not_exercise['대분류'] == '연령별']
print(not_ex_age)
# 원본 유지하고 복사
not_ex_gender = not_exercise[not_exercise['대분류'] == '성별'].copy()
# print(not_ex_gender)
# 대분류 삭제
not_ex_gender.drop(columns = '대분류', inplace = True)
# not_ex_gender = not_ex_gender.drop(columns = '대분류')
# not_ex_gender
# not_ex_gender.set_index(
```

```python
import matplotlib.pyplot as plt
# fig(figure : 그래프를 그리리 공간) , ax(axis : 그 공간 중 내가 사용할 부분)
fig,ax = plt.subplots(1, 2)  # subplots() - 작은 그래프(1행 2열)
plt.rc('font', family='Malgun Gothic')
not_ex_gender['운동을 할 충분한 시간이 없어서'].plot.pie(ax=ax[0], # 첫번째 그래프에 텍스트 삽입 
                autopct='%.1f%%', startangle=90)
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')
not_ex_gender['운동을 싫어해서'].plot.pie(ax=ax[1],    # 두번째 그래프에 텍스트 삽입 
                autopct='%.1f%%', startangle=90)
ax[1].set_title('운동을 싫어해서')
ax[1].set_ylabel('')
```
