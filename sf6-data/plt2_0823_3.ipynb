###### 자료의 분포 상태를 직사각형 모양의 막대그래프로 나타낸 것으로 데이터의 빈도에 따라 높이가 결정됨
###### hist(데이터 리스트)

```python
import matplotlib.pyplot as plt
import random
dice = []
for i in range(20):
    dice.append(random.randint(1,6))
print(dice)
plt.hist(dice, bins = 6)
                # bin - 가로 항목의 수를 정해주는것
                        # 안하면 다이스가 안나온 항목은 항목에 없음
plt.show()
```

```python
import csv
with open('datas/temp_seoul.csv', 'r')as f:
    data = csv.reader(f)             # encoding = 'utf-8'를 추가하는 이유 : 파일을 못 읽어온다면 해당 구문을 추가할 
     # 최고기온만 리스트에 담기
    temp_high = []
    header = next(data)
    for row in data:
        if row[-1]:
            row[0] = row[0].replace('\t', '')
            if int(row[0].split('-')[0]) >= 1980:
               if row[0].split('-')[1] == '08':
                   temp_high.append(float(row[-1]))
plt.rc('font', family = 'Malgun Gothic')
plt.title('1980년 이후 서울의 8월 최고기온 분포')
plt.hist(temp_high,bins = 20, color = 'g')
plt.show()
print(len(temp_high))
```
