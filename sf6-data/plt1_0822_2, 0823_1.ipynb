#### 시각화 라이브러리 사용 matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
# 리스트를 한개만 사용시 y축 값으로 인식야여 작성된다
# plt.plot([1,2,3,4],[10,30,20,40])
        # 가로축 값, 세로축 값
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.title('Basic Graph', pad = 100, loc = 'right')
# pad - 여백 , loc(location) - 위치 서렂ㅇ
plt.plot([1,2,3,4],[90,80,85,70], color = 'k')

plt.xlabel('학생번호')
plt.ylabel('시험점수')

# 격자 형태
axis = {'both','x', 'y'}
# plt.grid() 
# 가로 세로 다 표시
plt.grid(True, axis = 'x')
# x축에 대응하는 선 생성
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('범례 색상 표시')
plt.plot([10,20,30,40], label = '체중',color='g')
plt.plot([40,30,20,10], label = '출산률',color ='r')
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('범례 색상 표시')
plt.plot([10,20,30,40],'b.', label = 'circle')
                        # 색과 모양
plt.plot([40,30,20,10],'r^', label = 'triangle')
plt.rc('font', family = 'Malgun Gothic')
# 한글을 그냥쓰면 오류 생성> 위 구문 사용하기
plt.legend()
plt.show()
```

```python
# 막대 그래프 0823_1
import matplotlib.pyplot as plt

# # x = [0,1,2,3,4,5]
# x = range(6)
# # 위 구문과 같은 뜻
# plt.bar(x,y)

# y = [1,2,3,5,6,7]
x = range(3)
y = [100,400,200]
plt.bar(x,y, color = ['r','g','b'], width = 0.6)

plt.xticks(x,['2018','2019','2020'])

plt.show
```

```python
# 가로 막대 그래프 0823_1
import matplotlib.pyplot as plt

x = range(3)
y = [100,400,200]
plt.barh(x,y, color = ['r','g','b'], height = 0.6)

plt.yticks(x,['2018','2019','2020'])

plt.show
```
