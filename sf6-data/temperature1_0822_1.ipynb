```python
import csv
import matplotlib.pyplot as plt
with open('datas/temp_seoul.csv', 'r')as f:
    data = csv.reader(f)   
    next(data)
    temp_high =[]
    for row in data:
        # if row[-1] != "":
        #     row[-1] = float(row[-1])
        #     if row[-1] > max_temp:
        #         max_temp = row[-1]  
        if row[-1]:
            row[0] = row[0].replace('\t', '')
            # 년도 앞에 공백 제거
            # row[0] = row[0].split('-')
            # 년도를 정수화 해서 비교
            if int(row[0].split('-')[0]) > 1920:
                temp_high.append(float(row[-1]))
plt.rc('font', family = 'Malgun Gothic')
plt.title('기상 관측 서울 기온 변화')
plt.plot(temp_high,color = 'r')
plt.show()
```
