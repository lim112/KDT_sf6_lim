### 내 생일에 서울의 최저기온과 최저기온

```python
import csv
import matplotlib.pyplot as plt
# temp_high = []
with open('datas/temp_seoul.csv', 'r')as f:
    data = csv.reader(f)             # encoding = 'utf-8'를 추가하는 이유 : 파일을 못 읽어온다면 해당 구문을 추가할 
     # 최고기온만 리스트에 담기
    # global temp_high
    header = next(data)

    for row in data:
       if row[-1]:
          row[0] = row[0].replace('\t', '')
          if int(row[0].split('-')[0]) >= 1980:
             if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '27':
                 temp_high.append(float(row[-1])) 
                 # 최저 기온
                 temp_row.append(float(row[2]))
                 
    print(temp_high[-1])
    print(temp_row[-1])
    plt.rc('font', family = 'Malgun Gothic')
    plt.title('내 생일의 서울의 최고, 최저 기온', pad = 20)
    print(len(temp_high))
    plt.xlabel('연도별 기온')
    plt.plot(temp_row, color = 'b', label = 'high')
    plt.plot(temp_high,color = 'r', label = 'low')
    plt.show()
```
