```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 크롬 드라이버 실행
driver = webdriver.Chrome()
driver.get('http://www.yookyoung.org/site/main/main.jsp')

# 페이지의 body 태그에서 모든 텍스트 가져오기
body_text = driver.find_element(By.TAG_NAME, 'body').text

# 텍스트를 줄 단위로 분리
body_list = body_text.split('\n')

# 카테고리별 카운트 딕셔너리
cutures_dic = {'음악': 0, '전시회': 0, '박물관': 0, '무용': 0, '전통': 0, '연극': 0, '영화': 0, '대중': 0, '운동': 0}

# 각 줄에서 카테고리별 단어를 검색하여 카운트 증가
for body in body_list:
    for category in cutures_dic:
        if category in body:
            cutures_dic[category] += 1

# "음악" 카테고리의 결과 출력
print(f"음악: {cutures_dic['음악']}")
print(f"전시회: {cutures_dic['전시회']}")
print(f"박물관: {cutures_dic['박물관']}")
print(f"대중: {cutures_dic['대중']}")
print(f"무용: {cutures_dic['무용']}")
print(f"연극: {cutures_dic['연극']}")
print(f"영화: {cutures_dic['영화']}")
print(f"운동: {cutures_dic['운동']}")
print(f"전통: {cutures_dic['전통']}")

# 드라이버 종료
driver.quit()
```

```python
import pandas as pd

# 서울의 구 이름들
addresses = ['중랑구','강동','강북','강서','관악','광진','구로','금천','노원','도봉',
             '동대문','동작','마포','서대문','서초','성동','송파','양천','영등포',
             '용산','은평','종로']

# 빈 데이터프레임 생성: index는 category, columns는 구
combined_data = pd.DataFrame(index=category)

# 각 구에 대한 CSV 파일을 읽어서 구 이름을 열로 추가
for address in addresses:
    data = pd.read_csv(f'category_counts_{address}.csv')
    combined_data[address] = data['빈도 수'].values

# 'count_data_total.csv' 파일로 저장
combined_data.to_csv('count_data_total.csv', encoding='utf-8-sig')
```
