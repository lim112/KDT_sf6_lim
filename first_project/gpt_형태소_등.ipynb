```python
import requests
from bs4 import BeautifulSoup
from konlpy.tag import Kkma
from collections import Counter

# KoNLPy의 Kkma 분석기 로드
kkma = Kkma()

# 웹 페이지의 URL
url = 'https://www.museum.go.kr/site/main/home'

# 웹 페이지 요청 및 파싱
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# 페이지에서 텍스트 추출
text = soup.get_text()

# 텍스트를 단어 단위로 토큰화 (Kkma 사용)
def tokenize_text(text):
    return kkma.nouns(text)  # 명사만 추출

words = tokenize_text(text)

# 관심 있는 카테고리 리스트
categories = [
    '전시회', '박물관', '음악', '무용', '전통예술', 
    '연극공연', '영화관람', '대중공연'
]

# 카테고리와 관련된 텍스트 비율 계산
def calculate_category_proportions(words, categories):
    total_words = len(words)
    category_counts = Counter(word for word in words if word in categories)
    
    proportions = {category: category_counts.get(category, 0) / total_words for category in categories}
    
    return proportions

# 카테고리 비율 계산
category_proportions = calculate_category_proportions(words, categories)

# 결과 출력
print("카테고리 비율:")
for category, proportion in category_proportions.items():
    print(f"{category}: {proportion:.4f}")

```

```python
from pykospacing import Spacing
import requests
from bs4 import BeautifulSoup
from collections import Counter

# PyKoSpacing으로 띄어쓰기 교정
spacing = Spacing()

# 웹 페이지의 URL
url = '웹 페이지 URL을 여기에 넣으세요'

# 웹 페이지 요청 및 파싱
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# 페이지에서 텍스트 추출
text = soup.get_text()

# 띄어쓰기 교정
corrected_text = spacing(text)

# 텍스트를 단어 단위로 토큰화 (공백 기준으로 분리)
words = corrected_text.split()

# 관심 있는 카테고리 리스트
categories = [
    '전시회', '박물관', '음악 및 무용 발표회', '전통 예술 공연', 
    '연극 공연', '영화 관람', '대중 공연', '운동 경기'
]

# 카테고리와 관련된 텍스트 비율 계산
def calculate_category_proportions(words, categories):
    total_words = len(words)
    
    if total_words == 0:
        return {category: 0 for category in categories}  # 단어가 없는 경우 모든 비율을 0으로 설정
    
    category_counts = Counter(word for word in words if word in categories)
    
    proportions = {category: category_counts.get(category, 0) / total_words for category in categories}
    
    return proportions

# 카테고리 비율 계산
category_proportions = calculate_category_proportions(words, categories)

# 결과 출력
print("카테고리 비율:")
for category, proportion in category_proportions.items():
    print(f"{category}: {proportion:.4f}")
```
