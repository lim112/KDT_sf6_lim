```python
import pandas as pd
addresses = ['','','','','','','','','','',
            '','','','','','','','','','',
            '','','','','']
for address in addresses:
    #  CSV  
    df = pd.read_csv(f'count_{address}.csv')

    #     
    results = []

    #   
    for index, row in df.iterrows():
        #    (: ''   )
        categories = ['', '', '   ', '  ', ' ', ' ', ' ', ' ']
        scores = row[categories]
        
        #      
        max_category = scores.idxmax()
        max_value = scores.max()
        
        #   
        results.append({
            '  ': max_category,
            '': max_value
        })

    #    
    results_df = pd.DataFrame(results)

    #      ( )
    df = pd.concat([df, results_df], axis=1)

    #   CSV  
    df.to_csv(f'category_count{address}.csv', encoding='utf-8-sig', index=False)
```

```python
import pandas as pd

addresses = ['','','','','','','','','',
            '','','','','','','','','',
            '','','','']
categories_list = ['', '', '   ', '  ',
            ' ', ' ', ' ', ' ']

# for address in addresses:
    #  CSV  
df = pd.read_csv(f'count_.csv')

#     
categories_count = {category: 0 for category in categories_list}

# '  '    
for category in df['  ']:
    if category in categories_count:
        categories_count[category] += 1

#      
count_df = pd.DataFrame(list(categories_count.items()), columns=['', ' '])

#        
df['  '] = df['  '].map(categories_count)

#   CSV  
df.to_csv(f'count_.csv', encoding='utf-8-sig', index=False)

#      ( )
count_df.to_csv(f'category_counts_.csv', encoding='utf-8-sig', index=False)
```
