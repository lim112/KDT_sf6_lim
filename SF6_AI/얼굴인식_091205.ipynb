```python
import cv2
img = cv2.imread('./source/siblings.jpg')
# img = cv2.imread('http://localhost:8888/files/SF6_AI/source/pic.png?_xsrf=2%7Cd4d4c7e2%7Cb3417b178838332a2731fc871eff0407%7C1725444674')
# img = cv2.imread('http://localhost:8888/lab/tree/SF6_AI/source/pic.png')

# 이미지를 흑백으로 변환
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 가중치 파일 경로
cascade_file = 'haarcascade_frontalface_alt.xml'

# 모델 불러오기
face_cascade = cv2.CascadeClassifier(cascade_file)

# 1.3 - 30%(비울을 반복적으로 줄여감), 4 : 최소 추가 검출 횟수
# detectMultiScale - 객체 탐지 메서드
faces = face_cascade.detectMultiScale(img_gray, 1.3,4) 

# 얼굴 인식 사각형(rectangle) - x(좌표), y(좌표), w(너비), h(높이)
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,0) ,3,minSize=(1, 1)) 
    
cv2.imshow('pic', img)
# 이미지 크기 조절 (50%로 축소)
scale_percent = 50  # 50% 크기로 축소
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
img_resized = cv2.resize(img, dim)

# 이미지 출력
plt.imshow(img_resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
