```python
# 뉴스 기사 이미지 준비
import cv2
import matplotlib.pyplot as plt
# 이미지를 흑백으로 불러오기
img = cv2.imread('./source/news.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap = 'gray')

plt.axis('off')

plt.show()
```

```python
# tesseract 환경 설정
from pytesseract import pytesseract

pytesseract.tesseract_cmd = "C:/Users/vvv/AppData/Local/Programs/Tesseract-OCR/tesseract.exe" # r'C:\Users\vvv\AppData\Local\Programs\Tesseract-OCR' 역슬래시와 슬래시의 차이

# 이미지를 텍스트로 추출
text = pytesseract.image_to_string(img, lang='kor+eng') # 한글과 영어를 인식하게
print(text)      
```

```python
import cv2
import matplotlib.pyplot as plt
from pytesseract import pytesseract

# 이미지를 흑백으로 불러오기
img = cv2.imread('source/sonata1.png', cv2.IMREAD_GRAYSCALE)

# # 이진 이미지로 변경
_, bin_img = cv2.threshold(img,25, 255, cv2.THRESH_BINARY)

# 이진화된 이미지 출력
plt.imshow(bin_img, cmap='gray')
plt.axis('off')
plt.show()

# Tesseract 환경 설정
pytesseract.tesseract_cmd = \
                "C:/Users/vvv/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"

# 이진화된 이미지에서 텍스트 추출
text = pytesseract.image_to_string(bin_img, lang='kor')  # 한글과 영어를 인식
print(text)
```
