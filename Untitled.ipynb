```python
!pip install konlpy
# 성공
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
import requests
from bs4 import BeautifulSoup
from collections import Counter

# KoNLPy의 Kkma 분석기 로드
kkma = Kkma()
hannanum = Hannanum()

# 웹 페이지의 URL
url = 'https://www.jungnangimc.or.kr/html/01010100'

# 웹 페이지 요청 및 파싱
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# 페이지에서 텍스트 추출
text = soup.get_text()

# 텍스트를 단어 단위로 토큰화 (Kkma 사용)
def tokenize_text(text):
    return kkma.nouns(text)  # 명사만 추출

words = tokenize_text(text)

# 관심 있는 카테고리 리스트
categories = [
    '전시회', '박물관', '음악 및 무용 발표회', '전통 예술 공연',
    '연극 공연', '영화 관람', '대중 공연', '운동 경기'
]

# 카테고리와 관련된 텍스트 비율 계산
def calculate_category_proportions(words, categories):
    total_words = len(words)

    if total_words == 0:
        return {category: 0 for category in categories}  # 단어가 없는 경우 모든 비율을 0으로 설정

    category_counts = Counter(word for word in words if word in categories)

    proportions = {category: category_counts.get(category, 0) / total_words for category in categories}

    return proportions

# 카테고리 비율 계산
category_proportions = calculate_category_proportions(words, categories)

# 결과 출력
print("카테고리 비율:")
for category, proportion in category_proportions.items():
    print(f"{category}: {proportion:.4f}")
```

```python
!pip install transformers
!pip install torch
import requests
from bs4 import BeautifulSoup
from transformers import BertTokenizer, BertModel
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# BERT 모델과 토크나이저 로드
model_name = 'bert-base-multilingual-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# 관심 있는 카테고리 리스트
categories = [
  '전시회', '박물관', '음악 및 무용 발표회', '전통 예술 공연',
  '연극 공연', '영화 관람', '대중 공연', '운동 경기'
]

def get_text_from_url(url):
  """주어진 URL에서 텍스트를 추출합니다."""
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  return soup.get_text()

def get_bert_embeddings(text):
  """텍스트의 BERT 임베딩을 계산합니다."""
  inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
  with torch.no_grad():
      outputs = model(**inputs)
  # [CLS] 토큰의 임베딩을 사용합니다.
  cls_embedding = outputs.last_hidden_state[:, 0, :].numpy()
  return cls_embedding

def calculate_similarity(text, categories):
  """텍스트와 카테고리 간의 유사도를 계산합니다."""
  text_embedding = get_bert_embeddings(text)
  category_embeddings = np.vstack([get_bert_embeddings(cat) for cat in categories])

  # Cosine similarity 계산
  similarities = cosine_similarity(text_embedding, category_embeddings)
  return similarities.flatten()

# 웹 페이지 URL
addresses = ['강동','강북','강서','관악','광진','구로','금천','노원','도봉',
          '동대문','동작','마포','서대문','서초','성동','송파','양천','영등포',
          '용산','은평','종로','중랑구']
# 결과를 저장할 DataFrame 생성
count = pd.DataFrame(columns=['문화시설명', '전시회', '박물관', '음악 및 무용 발표회', '전통 예술 공연',
                            '연극 공연', '영화 관람', '대중 공연', '운동 경기'])
# for address in addresses:
# CSV 파일을 읽어옵니다.
data = pd.read_excel('filtered_data_강북.xlsx')

# 데이터 처리
for i, row in data.iterrows():
  url = row['홈페이지']
  facility = row['문화시설명']

  try:
      # 텍스트 추출
      text = get_text_from_url(url)

      # 유사도 계산
      similarity_scores = calculate_similarity(text, categories)

      # 새 데이터를 DataFrame으로 만들어 concat 사용
      new_row = pd.DataFrame({
          '문화시설명': [facility],
          '전시회': [similarity_scores[0]],
          '박물관': [similarity_scores[1]],
          '음악 및 무용 발표회': [similarity_scores[2]],
          '전통 예술 공연': [similarity_scores[3]],
          '연극 공연': [similarity_scores[4]],
          '영화 관람': [similarity_scores[5]],
          '대중 공연': [similarity_scores[6]],
          '운동 경기': [similarity_scores[7]]
      })

      # count DataFrame에 행 추가
      count = pd.concat([count, new_row], ignore_index=True)

  except Exception as e:
      print(f"Error processing {facility} with URL {url}: {e}")

# 결과를 CSV 파일로 저장
count.to_csv(f'count_강북.csv', encoding='utf-8-sig', index=False)
```
