### 실습2 전자신문 메인 기사 크롤링

```python
# 전자신문 크롤링
import requests
from bs4 import BeautifulSoup
url = "https://www.etnews.com/20240819000135"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
title = soup.select_one("div.article_title")
print(title.text)

time = soup.select("time.date")
print(time[0].text)

# body = soup.select_one("div.article_txt")
# print(body.text.strip())

# body = soup.select("div.article_txt > p")
#                             # div class(.)가 atricle_txt인것을 찾아 p테그가 body가 된다
body = soup.find_all("div", class_ = "article_txt")
for content in body:
    print(content.text)
```

```python
import requests
from bs4 import BeautifulSoup

# 전자신문의 특정 기사 URL
url = "https://www.etnews.com/20240819000135"
response = requests.get(url)

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# 제목 가져오기
title = soup.find("div", class_="article_title")
print(title.text.strip())

# 시간 가져오기
time = soup.find("time", class_="date")
print(time.text.strip())

# 본문 가져오기
body = soup.find("div", class_="article_body")
print(body.text.strip())
```
